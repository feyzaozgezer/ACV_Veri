Tüm Tahakkuk verilerini birleştirip ( pd.concat ) toplam kayıt sayısını doğrulayın

```python
# Tüm tahakkuk veri çerçevelerini birleştirir
df_all_tahakkuk = pd.concat([df_tahakkuk, df_tahakkuk_1, df_tahakkuk_2], ignore_index=True)

# Birleştirilmiş veri çerçevesinin toplam kayıt sayısını gösterir
print(f"Birleştirilmiş tüm Tahakkuk veri çerçevesinin kayıt sayısı: {df_all_tahakkuk.shape[0]}")

# Tek tek veri çerçevelerinin kayıt sayılarını toplayarak doğrulama yapar.
total_records_individual = df_tahakkuk.shape[0] + df_tahakkuk_1.shape[0] + df_tahakkuk_2.shape[0]
print(f"Tek tek Tahakkuk veri çerçevelerinin toplam kayıt sayısı: {total_records_individual}")

if df_all_tahakkuk.shape[0] == total_records_individual:
    print("Toplam kayıt sayısı doğrulandı: Birleştirilmiş veri çerçevesi ve bireysel veri çerçevelerinin toplam kayıt sayısı eşleşiyor.")
else:
    print("HATA: Kayıt sayısı doğrulanamadı. Birleştirilmiş ve bireysel veri çerçevelerinin toplam kayıt sayısı eşleşmiyor.")
```

### 'kwh' sütunundaki eksik değerleri tespit etme

```python
missing_kwh_count = df_all_tahakkuk['kwh'].isnull().sum()
print(f"'kwh' sütununda eksik değer sayısı: {missing_kwh_count}")

if missing_kwh_count > 0:
    print("Eksik değerlere sahip satırlar:")
    display(df_all_tahakk[df_all_tahakk['kwh'].isnull()].head())
```

### 'kwh' sütunundaki aşırı uç değerleri tespit etme (Kutu grafiği ile görselleştirme)

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(y=df_all_tahakkuk['kwh'])
plt.title('KWH Değerlerinin Kutu Grafiği')
plt.ylabel('KWH')
plt.show()

# İsteğe bağlı: IQR yöntemini kullanarak sayısal olarak aşırı uçları belirleyelim
Q1 = df_all_tahakkuk['kwh'].quantile(0.25)
Q3 = df_all_tahakkuk['kwh'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_kwh = df_all_tahakkuk[(df_all_tahakkuk['kwh'] < lower_bound) | (df_all_tahakkuk['kwh'] > upper_bound)]

print(f"'kwh' sütununda aşırı uç olarak tespit edilen değer sayısı (IQR yöntemi): {outliers_kwh.shape[0]}")
if outliers_kwh.shape[0] > 0:
    print("Aşırı uç değerlere sahip satırlar (ilk 5):")
    display(outliers_kwh.head())
```
